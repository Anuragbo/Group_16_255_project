## Section 5: Correlation Analysis

### Visualization 2: Correlation Matrix Heatmap

This heatmap visualizes the strength and direction of linear relationships between numeric features.

```python
# Display Correlation Matrix
img2 = Image('02_correlation_matrix.png')
display(img2)

# Print correlation matrix as table
print("\nCorrelation Matrix Values:")
print("="*80)
numeric_cols = df.select_dtypes(include=[np.number]).columns
corr_matrix = df[numeric_cols].corr()
print(corr_matrix.round(4))
```

### Key Insights from Correlation Analysis:

**Interpretation of Correlation Matrix:**

1. **Senior Citizen & Tenure (Correlation: 0.0166)**
   - Almost zero correlation = completely independent
   - Age and customer loyalty are NOT related
   - Senior citizens don't necessarily stay longer than younger customers

2. **Monthly Charges & Senior Citizen (Correlation: 0.2202)**
   - Weak positive correlation
   - Senior citizens tend to pay slightly more on average
   - Could indicate different service preferences or specific pricing

3. **Monthly Charges & Tenure (Correlation: 0.2479)**
   - Weak positive correlation (strongest in the matrix)
   - Longer-tenured customers tend to pay slightly more
   - May indicate service bundle growth over time or inflation adjustments

**Critical Finding**: All correlations are weak (< 0.25), suggesting:
- **These numeric features are relatively independent**
- **Churn is likely driven more by categorical factors** (contract type, service adoption, payment method)
- Need to focus on service features and contract terms for predictive modeling

## Section 7: Examine Relationships - Target Variable Analysis

### Visualization 4: Churn Distribution and Analysis

Understanding the distribution of our target variable is critical for modeling and business strategy.

```python
# Display Churn Analysis
img4 = Image('04_churn_analysis.png')
display(img4)

# Print churn statistics
print("\nChurn Distribution:")
print("="*80)
churn_counts = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100
print(f"No Churn:  {churn_counts['No']:,} customers ({churn_pct['No']:.2f}%)")
print(f"Churned:   {churn_counts['Yes']:,} customers ({churn_pct['Yes']:.2f}%)")
print(f"\nChurn Rate: {churn_pct['Yes']:.2f}%")
```

### Key Insights from Churn Analysis:

**Target Variable Distribution:**
- **Non-Churned**: 5,174 customers (73.46%) - Retained customers
- **Churned**: 1,869 customers (26.54%) - Customers who left

**Model Implications:**
1. **Class Imbalance**: 3:1 ratio (retained:churned) is moderate but notable
   - Recommendation: Use stratified sampling in cross-validation
   - Consider class weights or SMOTE oversampling during modeling
   - Use F1-Score, Precision, Recall, and ROC-AUC as evaluation metrics (not just accuracy)

2. **Churn Rate of 26.54% is Significant**:
   - Industry context: Telecom average churn is typically 20-30% annually
   - This company is within normal range but with room for improvement
   - Each 1% reduction in churn would retain 70+ additional customers annually

3. **Business Impact**:
   - With ~7,000 customer base, even small improvements in retention have major financial impact
   - Focus should be on high-value customer retention strategies
   - Identify early warning signs of churn for proactive intervention

**Next Steps for Modeling:**
- Analyze churn distribution by contract type, service adoption, and payment method
- Build predictive model to identify high-risk customers
- Develop targeted retention strategies for at-risk segments
